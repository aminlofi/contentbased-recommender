```python
import pandas as pd
import re 
```

```python
# load data from csv file (dataset)
df = pd.read_csv("C:/Users/Amin/project-amin/web scraping/sephoramy_webscrape_dataset_raw.csv")
```

```python
# returns 5 row of the dataframe
df.head() 
```

```python
# return data type of each object and memory usage
df.info()
```

```python
# drop all the rows data with missing values
df = df.dropna()
df.info()
```

```python
# reset back the index number after changes earlier, drop the old index column and save the updated changes 
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
# category, url and brand have nothing to be clean, hence proceed to name
# cleaning name column
# since there exist multiple duplicate products, we can drop this rows by identifying them from their name

# drop duplicate product name 
df_2 = df['name'].drop_duplicates()

# resetting index after dropping 
df = df.loc[df_2.index, :].reset_index()

# drop index 
df.drop(['index'], axis = 1, inplace = True)
```

```python
# cleaning price column
# convert RM price string to numerical

pattern = re.compile(r"(\d+).\d+")
for i in range (len(df)):
    df['price'][i] = re.findall(pattern, df['price'][i])[0]
    
df['price'] = pd.to_numeric(df['price'])
```

```python
df.head()
```

```python
# cleaning skin type column
# remove skin type word to leave only the skin type variables into the dataset

pattern = 'Skin Type: '
for i in range(len(df)):
    df['skin_type'][i] = re.sub(pattern, '', df['skin_type'][i])
```

```python
df.head()
```

```python
# convert skin type variables into dummy indicator  

df_3 = df['skin_type'].str.get_dummies(sep=', ')
df_4 = df.join(df_3).drop('skin_type', axis = 1)
```

```python
df_4.head()
```

```python
df_4.to_csv('sephoramy_webscrape_dataset_clean.csv', encoding = 'utf-8-sig', index = False)
```
